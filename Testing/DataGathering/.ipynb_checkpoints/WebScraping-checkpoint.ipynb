```python
import requests
```

```python
URL = "https://ned.ipac.caltech.edu/cgi-bin/OBJatt?delimiter=bar&nondb=row_count&crosid=objname&position=ra%2Cdec&enotes=objnote&position=pretype&position=z&position=zflag&attdat_CON=M&attdat_CON=S&attdat=attned&gphotoms=q_value&gphotoms=q_unc&gphotoms=ned_value&gphotoms=ned_unc&diamdat=ned_maj_dia&distance=avg&distance=stddev_samp&M=1010"
```

```python
r = requests.get(url=URL)
```

```python
r
```

```python
import bs4
```

https://stackoverflow.com/questions/50633050/scrape-tables-into-dataframe-with-beautifulsoup

https://www.geeksforgeeks.org/how-to-remove-tags-using-beautifulsoup-in-python/

```python
soup = bs4.BeautifulSoup(r.text, "lxml")
```

```python
table = soup.find("table")
table_rows = table.find("pre")
```

```python
table_rows.find("strong").decompose()
```

```python
table_rows = table_rows.get_text().splitlines()[1:]
```

```python
def extract_data(string: str) -> list[str]:
    return [*map(lambda item: item.strip(), string.split("|"))]
table_rows = [*map(extract_data, table_rows)]
```

```python
table_rows
```

```python
import pandas as pd
```

```python
df_cols = [
    "Number",
    "Space",
    "Name",
    "RA",
    "Dec",
    "Object Type",
    "Redshift",
    "z",
    "Morphology",
    "Activity",
    "Number2"
]
df = pd.DataFrame(columns=df_cols)
```

```python
for row in table_rows:
    df_row = {col: val for (col, val) in zip(df_cols, row)}
    df.loc[len(df)] = df_row
```
